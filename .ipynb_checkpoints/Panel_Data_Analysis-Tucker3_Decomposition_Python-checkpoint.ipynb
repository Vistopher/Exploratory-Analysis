### trying tucker3 decomposition using python

define the function of tucker3 decomposition

```python
import numpy as np
import tensorly as tl
from tensorly.decomposition import tucker
import ast

def load_tensor_from_file(file_path):
    with open(file_path, 'r') as file:
        contents = file.read()
    tensor_data = np.array(ast.literal_eval(contents))
    return tensor_data

def perform_tucker_decomposition(tensor_data, ranks):
    tensor = tl.tensor(tensor_data)
    core, factors = tucker(tensor, rank=ranks)
    return core, factors

file_path = 'tensor.txt'
tensor_data = load_tensor_from_file(file_path)
print(tensor_data.shape)
```

## Step 2:
### Conduct Tucker3 analysis with $X_{new}(n\times m \times r)$ .The component load constraint of the three modes is orthogonal, and the component decomposition dimensions P,Q,S of the model are calculated according to the $\overline{R}^2$ under the premise of ensuring the interpretability of the model

Tucker3 model:
$$
x_{ijt} = \sum_{p=1}^{P} \sum_{q=1}^{Q} \sum_{s=1}^{S} b_{jq}c_{ts}g_{pqs} + e_{ijt}                
$$
and in a matrix format:
$$
X_{a} = AG_{a}(C'\otimes B') + E_{a} 
$$

```python
from tensorly import tucker_to_tensor, norm

best_fit = None
best_percent = -np.inf
mean_tensor = tl.mean(tensor_data)

n = 31
m = 4
r = 19

print(tensor_data.shape[0])
print(tensor_data.shape[1])
print(tensor_data.shape[2])


for p in range(1, min(n, tensor_data.shape[1])+1):
    for q in range(1, min(m, tensor_data.shape[2])+1):
        for s in range(1, min(r, tensor_data.shape[0])+1):
            tucker_result = tucker(tensor_data, rank=[p, q, s], n_iter_max=25, tol=1e-05)
            core = tucker_result.core
            factors = tucker_result.factors
            reconstruction = tucker_to_tensor(tucker_result)
            resid_2 = norm(tensor_data - reconstruction, 2)**2
            total_var = norm(tensor_data - mean_tensor, 2)**2
            R_2 = 1 - resid_2 / total_var
            N = np.prod(tensor_data.shape)
            k = sum(factor.shape[0] for factor in factors) + np.prod(core.shape) - 1
            R_2_adj = 1 - ((1 - R_2) * (N - 1) / (N - k - 1))
            
            if isinstance(R_2_adj, np.float64) and not np.isnan(R_2_adj):
                if R_2_adj > best_percent:
                    best_percent = R_2_adj
                    best_fit = tucker_result

print(best_percent)
if best_fit is not None:
    print("Best core shape:", best_fit.core.shape)
    print("Best factors shapes:", [factor.shape for factor in best_fit.factors])
```

## Step 3:

### Calculate the combined score of the "variable-time" mode on the p-th component of the individual mode $score_{pjt}$

$$
score_{pjt} = \sum_{q=1}^{Q} \sum_{s=1}^{S} b_{jq}c_{ts}g_{pqs}
$$

```python
core = best_fit.core
factors = best_fit.factors

P = factors[1].shape[1]  
Q = factors[2].shape[1]  
S = factors[0].shape[1] 

p = 0  
j = 0
t = 0
score_pjt = np.zeros(P)
print(score_pjt)

for p in range(P):
    for q in range(Q):
        for s in range(S):
            score_pjt[p] += factors[2][j, q] * factors[0][t, s] * core[s, p, q]    
    print(f"Score for factor {p+1}: {score_pjt[p]}")
print("Scores:", score_pjt)
```

## Step 4:
### Calculate the Minkowski distance between two region slices:$X_{i1..} (m \times r)$ and $X_{i2..} (m \times r)$

Given the panel data $X (n \times m \times r)$ where the two horizontal slice matrices are $X_{i1..} (m \times r)$ and $X_{i2..} (m \times r)$，$i1, i2 = 1, 2, ..., n$，and both matrices are observations of the same $m$ variables over $r$ periods, the "individual" distance $d_{i1i2}^A$ between $i1$ and $i2$ is defined as:

$$
d_{i1i2}^A = \left( \sum_{p=1}^{P} w_p(a_{i1p} - a_{i2p})^h \right)^{1/h} = \left( \sum_{p=1}^{P} \sum_{q=1}^{Q} \sum_{s=1}^{S} b_{jq}c_{ts}g_{pqs}(a_{i1p} - a_{i2p})^h \right)^{1/h}
$$

Here，$w_p = score_{pjt}$，and in this article $d_{i1i2}^A$  is defined based on a form of Minkowski distance, which can be adjusted according to different research purposes and data types. Furthermore, based on this equation, the distances between "variable modes" and "time modes" can also be articulated, thus allowing for separate clustering of variables and time.

We set h to 3 at this time, and we can change it if we want more experiments.

```python
h = 3
sum_power_h = 0
A = factors[1]
n,P = A.shape
distance_d_i1_i2_A = np.empty((n,n))

for i in range(n): 
    for j in range(n):
        for p in range(P):
            sum_power_h += score_pjt[p] * (np.abs(A[i,p] - A[j, p])**h)
        distance_d_i1_i2_A[i,j] = sum_power_h**(1/h)  
        sum_power_h = 0   
print("Weighted Minkowski Distance:")
print(distance_d_i1_i2_A)
```

## Step 5:
### Perform traditional two-way data clustering analysis on the three-way "individual" distance matrix distance_d_i1_i2_A.

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
from scipy.spatial.distance import squareform

# Convert the distance matrix into a condensed distance matrix. 
condensed_distance_d_i1_i2_A = squareform(distance_d_i1_i2_A)

#  Use the Ward linkage criterion for hierarchical clustering.
Z = linkage(condensed_distance_d_i1_i2_A, method='ward')

# Visualize with a dendrogram.
plt.figure(figsize=(35, 15))
dendrogram(Z)

plt.title('Cluster Analysis of Solid Waste Treatment in 31 Provinces and Regions of China from 2003 to 2021 using PCT3 Method; Model - Tucker3(6,4,3), Adjusted R-squared -- 98.16%')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.savefig('hierarchical_clustering.png', dpi = 700)  
plt.show()
```
