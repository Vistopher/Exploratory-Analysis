### Data description: 
Now we use the full panel data for analysis. To reflect how solid waste is treated and utilized from 2003-2021 in 31 Regions in China, I collected data from Chinese Statistical Yearbook. There is the panal dataset called'Disposal and Utilization of Industrial Solid Wastes by Region' in the Resource and Environment section most of the years between 2003-2021. 

#### Data issues:
The original datasets spanning multiple years face issues of data gaps and inconsistent indicators. 

Firstly, there are data gaps. Relevant data for all provinces and cities for the years 2007-2010 and 2018-2019 are missing or not publicly available. In some provinces and cities, especially Tibet, there is a severe lack of detailed sub-indicators related to hazardous solid waste.

Secondly, there is the issue of inconsistent indicators. Starting from 2011, the publicly available statistical indicators have become more detailed, such as splitting "Industrial Solid Waste Generation" before 2007 into "General Industrial Solid Waste Generation" and "Hazardous Industrial Solid Waste Generation." Additionally, from 2017 onwards, "Hazardous Waste Utilization Quantity" and "Hazardous Waste Disposal Quantity" were merged into a single indicator.


#### Measures:
Regarding how to handle missing data, I initially had two methods that could handle datasets with missing values. However, for the sake of simplifying the analysis, I chose to first fill in the data. Upon careful examination, I noticed that in some provinces and cities, the missing sub-indicators often corresponded to very low values of the overall indicators, so I decided to disregard them.

Regarding the issue of inconsistent indicators across years, in order to make full use of the data from each year and to showcase the temporal dimension of the indicators as much as possible, I used the total indicators from 2003-2006 as the baseline and merged the sub-indicators from 2011-2021. This ultimately resulted in four total indicators. From 2017 onwards, due to the merger of 'Hazardous Waste Utilization Quantity' and 'Hazardous Waste Disposal Quantity' into a single indicator, for the sake of convenience, I evenly divided them and separately included 'General Industrial Solid Waste Utilization Quantity' and 'General Industrial Solid Waste Disposal Quantity.'"

#### Final indexes:
The four summarized indexes are Output, Utilization, Storage and Disposition. 
In which, Output refers to the industrial solid waste generation, which is the sum of General Industrial Solid Waste Generation and Hazardous Waste Generation. Utilization refers to the comprehensive utilization of industrial solid waste, which is the sum of General Industrial Solid Waste Comprehensive Utilization and Hazardous Waste Comprehensive Utilization. Storage refers to the industrial solid waste storage, which is the sum of General Industrial Solid Waste Storage and Hazardous Waste Storage. The Disposal refers to the industrial solid waste disposal, which is the sum of General Industrial Solid Waste Disposal and Hazardous Waste Disposal.

* **Extraction of Time-Series Characteristics from Panel Data: Comprehensive 'Absolute Quantity' Features (AQF), Comprehensive 'Volatility' Features (VF), Comprehensive 'Skewness' Characteristics (SCF), Comprehensive 'Kurtosis' Characteristics (KCF), Comprehensive 'Trend' Features (TF).**

* #### **Definitions**
For a panel data set $x_{ik}(t)$, assuming there are $N$ samples, each with $P$ indicators over $T$ periods, the following definitions are made:

1. **Absolute Quantity Feature (AQF) of sample $i$ for indicator $k$**, denoted as $AQF(F_{ik})$:
   $$AQF(F_{ik}) = \frac{1}{T} \sum_{t=1}^{T} x_{ik}(t)$$
   $AQF(F_{ik})$ is essentially the mean of the $k^{th}$ indicator of sample $i$ over the total period $T$. This feature reflects the absolute level of development of the individual sample over the entire analysis period.

2. **Variance Feature (VF) of sample $i$ for indicator $k$**, denoted as $VF(F_{ik})$:
   $$VF(F_{ik}) = \frac{1}{T-1} \sum_{t=1}^{T} (x_{ik}(t) - \overline{x_{ik}})^2$$
   where $\overline{x_{ik}}$ is the mean of the $k^{th}$ indicator of sample $i$ over the total period $T$. $VF(F_{ik})$ characterizes the degree of fluctuation in the indicator values over time for individual $i$ during period $T$.

3. **Skewness Coefficient Feature (SCF) of sample $i$ for indicator $k$**, denoted as $SCF(F_{ik})$:
   $$SCF(F_{ik}) = \frac{\sum_{t=1}^{T} (x_{ik}(t) - \overline{x_{ik}})^3}{T (\sigma_{ik})^3}$$
   where $\sigma_{ik}$ is the standard deviation of the $k^{th}$ indicator for sample $i$ over period $T$. $SCF(F_{ik})$ measures whether the indicator values are symmetric over the entire period $T$. A negative $SCF(F_{ik})$ indicates that most values, including the median, are on the right side of the mean, and vice versa.

4. **Kurtosis Coefficient Feature (KCF) of sample $i$ for indicator $k$**, denoted as $KCF(F_{ik})$:
   $$KCF(F_{ik}) = \frac{\sum_{t=1}^{T} (x_{ik}(t) - \overline{x_{ik}})^4}{T (\sigma_{ik})^4} - 3 $$
   This feature represents the concentration degree or the peakedness of the distribution curve of indicator values for individual $i$ during period $T$. A $KCF(F_{ik})$ greater than 0 indicates a more concentrated distribution around the mean than a normal distribution, and vice versa if $KCF(F_{ik})$ is less than 0.

5. **Trend Feature (TF) of sample $i$ for indicator $k$**, denoted as $TF(F_{ik})$:
   $$ TF(F_{ik}) = \sum_{t=1}^{T} \left( x_{ik}(t) - \overline{x_{ik}} \right) \left( \frac{t - \frac{T}{2}}{\frac{T}{2}} \right)^2 $$
   $TF(F_{ik})$ describes the long-term change (monotonic characteristic) of the indicator values. If two indicators change in the same direction over time, the more synchronized their changes, the more similar they are.


```r
# Define a function for Trend Feature (TF)
compute_TF <- function(x, t_seq) {
  T <- length(x)
  numerator <- sum((x - mean(x, na.rm = TRUE)) * (t_seq - T/2))
  denominator <- sum((t_seq - T/2)^2)
  return(numerator / denominator)
}

library(purrr)
library(moments)  # for skewness() and kurtosis()

# Define a function for feature extraction
extract_features <- function(data, column_name) {
  features <- data %>%
    group_by(Region) %>%
        summarise(
          AQF = mean(.data[[column_name]]),
          VF = sd(.data[[column_name]]),
          SCF = skewness(.data[[column_name]]),
          KCF = kurtosis(.data[[column_name]]),
          TF = compute_TF(.data[[column_name]], 1:length(.data[[column_name]]))
        )
  #  Rename columns
  colnames(features) <- c("Region", paste0(column_name, "_AQF"), paste0(column_name, "_VF"), paste0(column_name, "_SCF"), paste0(column_name, "_KCF"), paste0(column_name, "_TF"))
  return(features)
}

# Define a list of indicators
indicators <- c("Output", "Utilization", "Storage", "Disposition")

# Use map_dfr() to extract features for all indicators
df1_features <- map_dfc(indicators, ~extract_features(df1_standardized, .x))
df1_features <- df1_features %>%
    rename(Region = 1) %>%   # Rename the first column as "Region"
        select(Region, contains("Output"),contains("Utilization"),contains("Storage"),contains("Disposition"))  # 选择Region列和其它指标列

# Check the results-the feature vector
df1_features
```

* **Standardize the feature vector by defining a standardization function.**

* #### **Definition**
6: **Feature Vector of sample i**, denoted as $M_i$:

$$M_i = (M_{i1}, M_{i2}, \ldots, M_{ik}, \ldots, M_{iP}) = [\omega_1 AQF(M_{i1}), \omega_2 VF(M_{i1}), \omega_3 SCF(M_{i1}), \omega_4 KCF(M_{i1}), \omega_5 TF(M_{i1}), \ldots, \omega_1 AQF(M_{ik}), \omega_2 VF(M_{ik}), \omega_3 SCF(M_{ik}), \omega_4 KCF(M_{ik}), \omega_5 TF(M_{ik}), \ldots, \omega_1 AQF(M_{iP}), \omega_2 VF(M_{iP}), \omega_3 SCF(M_{iP}), \omega_4 KCF(M_{iP}), \omega_5 TF(M_{iP})] $$

Here, $$M_{ik} = [\omega_1 AQF(M_{ik}), \omega_2 VF(M_{ik}), \omega_3 SCF(M_{ik}), \omega_4 KCF(M_{ik}), \omega_5 TF(M_{ik})]$$, is the feature vector for the $k^{th}$indicator of sample i.

$AQF(M_{ik})$, $VF(M_{ik})$, $SCF(M_{ik})$, $KCF(M_{ik})$, and $TF(M_{ik})$ are the normalized values of $AQF(F_{ik})$, $VF(F_{ik})$, $SCF(F_{ik})$, $KCF(F_{ik})$, and $TF(F_{ik})$, respectively. The normalization is done using a mean-based method, with the specific formula as follows:

$$AQF(M_{ik}) = \frac{AQF(F_{ik}) + VF(F_{ik}) + SCF(F_{ik}) + KCF(F_{ik}) + TF(F_{ik})}{5}$$

The definitions for $VF(M_{ik})$, $SCF(M_{ik})$, $KCF(M_{ik})$, and $TF(M_{ik})$ are similar to that of $AQF(M_{ik})$. Additionally, $\omega_1$, $\omega_2$, $\omega_3$, $\omega_4$, $\omega_5$ represent the weights of the five features, and they satisfy $\omega_1$ + $\omega_2$ + $\omega_3$ + $\omega_4$ + $\omega_5$ = 1. The weight coefficients can be subjectively assigned or objectively determined based on the actual situation and focus of the research problem.


```r
# Define a Standarlization Function
standardize <- function(data, prefix) {
  cols <- grep(prefix, names(data), value=TRUE)  
  sum_cols <- rowSums(data[cols])                
  sapply(cols, function(col) data[[col]] / sum_cols) 
}

# Use an lapply loop to standardize each prefix
prefixes <- c("Output", "Utilization", "Storage", "Disposition")
standardized_data <- do.call(cbind, lapply(prefixes, function(prefix) standardize(df1_features, prefix)))

# Add the Region column back to the standardized data
df1_features_standardized <- cbind(df1_features['Region'], standardized_data)
df1_features_standardized

# Retrieve the data for the first row's first five columns
row_values <- df1_features_standardized[1, 2:6]

# Calculate the sum of these five columns
sum_value <- sum(row_values, na.rm = TRUE)  
sum_value
```

* **The weight coefficients can be subjectively determined based on the focus. However, here I intend to write an example of the Analytic Hierarchy Process (AHP) to determine the weights of each feature, as this method can be widely applied in fields requiring authoritative expert opinions. Further, I have also written a Bayesian-adjusted Analytic Hierarchy Process. Due to limited technical capabilities, I was unable to completely encapsulate it to form a comprehensive example for use. At this moment, I have input the judgment matrix. The judgment matrix here does not have practical significance; it only serves as a progression of this method.**

* **Define a function for "computing the Analytic Hierarchy Process (AHP) of a hierarchical structure." The input consists of one: judgment matrix, and the output consists of four: weight vector (stored as a vector), maximum eigenvalue (stored as a floating-point number), consistency index (stored as a floating-point number), and consistency ratio (stored as a floating-point number).**

```r
aAHP <- function(X){
    W <- apply(X, 1, prod)
    n <- length(W)  
    W <- W^(1/n)
    W <- W / sum(W)
    lambda_max <- mean(X %*% W / W)
    CI <- (lambda_max - n) / (n - 1)
    RI <- c(0, 0, 0.58, 0.90, 1.12, 1.24, 1.32, 1.41, 1.45, 1.49, 1.51)
    CR <- CI / RI[n]
    return(list(W = W ,lambda_max = lambda_max, CI = CI, CR = CR))
}

# Call the function aAHP.
X <- matrix(c(1,1/2,3,4,
             2,1,7,5,
            1/4,1/7,1,1/2,
            1/3,1/5,2,1), nrow = 4,byrow = TRUE)
B1 <- matrix(c(1,1/2,3,4,3,
             2,1,7,5,5,
            1/4,1/7,1,1/2,1/3,
            1/3,1/5,2,1,1,
            1/3,1/5,3,1,1), nrow = 5,byrow = TRUE)
B2 <- B1
B3 <- B1
B4 <- B1

aAHP(X)
aAHP(B1)
aAHP(B2)
aAHP(B3)
aAHP(B4)
```

```r
A <- aAHP(X)$W
A
B <- cbind(aAHP(B1)$W,aAHP(B2)$W,aAHP(B3)$W,aAHP(B4)$W)
colnames(B)<-c("B1", "B2", "B3", "B4")
B
```

* **Obtain the final weight vector for each factor, W.**

```r
W <- B%*%A
W
```

* **Use the Bayesian formula to adjust the final composite weight vector obtained from AHP.**

```r
W_bayes<-vector("numeric", nrow(B))
for (i in 1:nrow(B)){
    W_bayes[i] = 0
    for (j in 1:length(A)){
    W_bayes[i] = W_bayes[i]+ B[i,j]^2*A[j]/W[i]
    }
}
W_bayes
```

* **Form a clustered dataset using the synthesized final weighted vector, adjusted through both the general and Bayesian formulas.**

```r
final_df1 <- df1_features_standardized %>%
  rowwise() %>%
      mutate(across(-Region, ~ .x * rep(W, length.out=length(.x))))
final_df1

final_df1_bayes <- df1_features_standardized %>%
  rowwise() %>%
      mutate(across(-Region, ~ .x * rep(W_bayes, length.out=length(.x))))
final_df1_bayes
```

* **Preliminarily determine the optimal K value through two methods: 1. Elbow method (Scree plot): Identifying the point of maximum curvature.2. Silhouette coefficient method: A value closer to 1 is considered better.**

```r
library(ggplot2)
final_df1_clustering <- final_df1 %>% select(-Region)
final_df1_bayes_clustering <- final_df1_bayes %>% select(-Region)

# Define a function for determining the optimal k using the elbow method.
Optimalk_elbow <- function(data, k.max){
    set.seed(123) 
    wcss <- map_dbl(1:k.max, ~kmeans(data, .x)$tot.withinss)

    tibble(k = 1:k.max, wcss = wcss) %>%
      ggplot(aes(x = k, y = wcss)) +
      geom_line(aes(group = 1), color = "blue") +
      geom_point(color = "red") +
      labs(title = "Elbow Method for Optimal k",
           x = "Number of clusters k",
           y = "Total Within Sum of Squares") +
      theme_minimal()
}

# Call the function
Optimalk_elbow(final_df1_clustering,10)
Optimalk_elbow(final_df1_bayes_clustering,10)
```

```r
library(cluster)
library(purrr)

# Define a function for determining the optimal k using the silhouette coefficient method.
Profile_coefficient <- function(data, k.max){
    #  Calculate the silhouette coefficient for each K value
    compute_silhouette <- function(k) {
      km.res <- kmeans(data, centers = k)
      sil_res <- silhouette(km.res$cluster, dist(data))
      mean(sil_res[, "sil_width"])
    }
    # compute the silhouette coefficient for each K value within the range from 1 to k.max,
    sil_values <- map_dbl(2:k.max, compute_silhouette)
    
    # plot the curve to show how the silhouette coefficient changes with K values.
    tibble(k = 2:k.max, silhouette = sil_values) %>%
      ggplot(aes(x = k, y = silhouette)) +
      geom_line(aes(group = 1), color = "blue") +
      geom_point(color = "red") +
      labs(title = "Silhouette Method for Optimal k",
           x = "Number of clusters k",
           y = "Average Silhouette Value") +
      theme_minimal()   
}
Profile_coefficient(final_df1_clustering,10)
Profile_coefficient(final_df1_bayes_clustering,10)
```
